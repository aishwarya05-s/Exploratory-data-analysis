```python
#Importing necessary libarabies

import pandas as pd  # for import file(csv,excel,etc)

import numpy as np # for numerical opt.
import seaborn as sns # adv. visulazation
import matplotlib.pyplot as plt  # basic visulaztion
import warnings
```

```python
#Importing dataset

df=pd.read_csv("Titanic_dataset.csv")
df.head()
```

```python
#Dropping unnecessary column

df=df.drop(['PassengerId'],axis=1)
```

```python
df.describe()
```

```python
#To get an overview of data

df.info()
```

```python
#finding null values

df.isnull().sum()
```

```python
#Mean imputation for continuous varaibale age

df['Age'].fillna(df['Age'].mean(), inplace=True)
```

```python
df
```

```python
#dropping column as it has maximum missing values
df.drop(['Cabin'],axis=1)
```

```python
#to count number of times unique row appears
df.value_counts()
```

```python
#plot of pairplot
sns.pairplot(df)
```

```python
#plot of heatmap

plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title("Spearman Correlation")
plt.show()
```

```python
#plot of histogram
df.hist(bins=10,figsize=(10,10))
plt.show()
```

```python
#plot of boxplot

fig=plt.figure(figsize=(21,15))

plt.subplot(3,3,1)
sns.boxplot(df['Age'])
plt.subplot(3,3,2)
sns.boxplot(df['SibSp'])
plt.subplot(3,3,3)
sns.boxplot(df['Parch'])
plt.subplot(3,3,4)
sns.boxplot(df['Fare'])
plt.subplot(3,3,5)
sns.boxplot(df['Cabin'])
plt.subplot(3,3,6)
sns.boxplot(df['Embarked'])

plt.show()
```
